```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initialisation
K = 10000
max_tosses = 6
total_success = 0
running_average_list = [] # for plotting
count_succes_list = [] # for plotting
total_variance = 0
running_average_nr_successes_list = []
total_count_success = 0
```

```python
# source : https://towardsdatascience.com/bernoulli-distribution-with-python-from-scratch-89fda3c822b
def bernoulli(p,size=1):
    """
    random variates
    """
    rvs = np.array([])
    for i in range(0,size):
        if np.random.rand() <= p:
            a=1
            rvs = np.append(rvs,a)
        else:
            a=0
            rvs = np.append(rvs,a)
    return rvs
```

```python
file1 = open("C:\\Users\\Milan\\Documents\\Master hir 1\\Sem2\\Simulation\\Assignments\\Assignment 1\\coin_toss_ex10.txt","w") # OPEN OUTPUT FILE
file1.write('Run' + '\t' + 'nr_successes' + '\t' + 'Running avg' + '\n') 
for k in range(1, K + 1):
    count_success = 0
    nr_tosses = 0
    toss = bernoulli(0.5)
    if toss == 0:
        current_toss = 0
    elif toss == 1:
        current_toss = 1
    nr_tosses += 1
    while True:
        toss = bernoulli(0.5)
        if toss == 0:
            previous_toss = current_toss
            current_toss = 0
        else:
            previous_toss = current_toss
            current_toss = 1
        nr_tosses += 1
        if previous_toss != current_toss:
            count_success += 1
        if not nr_tosses < max_tosses:
            break
    #keep count succes in a list for plotting
    count_succes_list.insert(k, count_success)
    if count_success == max_tosses - 1:
        total_success += 1

    #running average of the probability
    running_average = total_success/k 
    #keep it in a list for plotting
    running_average_list.insert(k, running_average)

    #build up the variance
    total_variance += count_success*count_success

    #running average number of successes
    total_count_success += count_success
    running_average_nr_successes = total_count_success/k
    running_average_nr_successes_list.insert(k,running_average_nr_successes)

    
    #print('Run', k, 'gain', gain,  'Running_avg', running_average) 
    file1.write('%d' % k + '\t' + '%d' % count_success + '\t' + '%f' % running_average + '\n' ) # PRINT RESULTS IN OUPTUT FILE
        

```

```python
simulations = list(range(1,10000 + 1))
```

```python
def plot():
        plt.figure(figsize = (12,4))
        plt.subplot(1,2,1)

        plt.xlabel('run')
        plt.ylabel('Probability of succes')
        plt.plot(simulations, running_average_list, 
                label='Running average Probability of success')
        plt.legend()

        plt.subplot(1,2,2)
        plt.xlabel('run')
        plt.ylabel('success')
        plt.plot(simulations, count_succes_list, 
                label='#succeses')
        plt.plot(simulations, running_average_nr_successes_list, 
                label = 'running average of number of successes')
        plt.legend()
plot()
```
